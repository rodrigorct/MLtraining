# Advertising - Project 

This Data Science project contains an artificial dataset for web advertisings to practice exploratory data analysis (EDA) and machine learning techniques.  Is the customer cliked or not in the publicity?

Note: This work is not sponsored by any advertising company and it is just one of an outcome of a fun project made using Machine Learning techniques. The project does not aim at the violation of any policies or privacy since the data on is publicly available.

This dataset contains the following fields:

* 'Daily Time Spent on Site': in minutes.
* 'Age': age of the customer.
* 'Area Income': average customer income in the region.
* 'Daily Internet Usage': Daily internet average usage in minutes.
* 'Ad topic line'.
* 'City'.
* 'Male'. 
* 'Country'.
* 'Timestamp'.
* 'Clicked on Ad'': 0 or 1 indicate wheter the customer clicked or not on the advertising.

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = df['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
